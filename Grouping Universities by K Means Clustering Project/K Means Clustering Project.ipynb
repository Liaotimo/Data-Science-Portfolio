
# Grouping Universities by K Means Clustering Project


In this project, k-means clustering will be used to group university enrollment data into two clusters. Results from the unsupervised learning task will then be compared to clustering by Private or Public label. Note that Private and Public label was not used in model training, and is just used for comparison.
This exercise uses a sample data set from Udemy's Data Science and Machine Learning Bootcamp series provided by [Pierian Data](http://www.pieriandata.com). 


The data contains 777 observations across 18 variables.

Description of variables:
* Private A factor with levels No and Yes indicating private or public university
* Apps Number of applications received
* Accept Number of applications accepted
* Enroll Number of new students enrolled
* Top10perc Pct. new students from top 10% of H.S. class
* Top25perc Pct. new students from top 25% of H.S. class
* F.Undergrad Number of fulltime undergraduates
* P.Undergrad Number of parttime undergraduates
* Outstate Out-of-state tuition
* Room.Board Room and board costs
* Books Estimated book costs
* Personal Estimated personal spending
* PhD Pct. of faculty with Ph.D.â€™s
* Terminal Pct. of faculty with terminal degree
* S.F.Ratio Student/faculty ratio
* perc.alumni Pct. alumni who donate
* Expend Instructional expenditure per student
* Grad.Rate Graduation rate

## Grouping data by k-means clustering

```python
from sklearn.cluster import KMeans
```

```python
# Instantiating a k-means model
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private', axis=1))
```

```python
# Look at cluster centers
kmeans.cluster_centers_

 
```

```python
cluster_zip = list(zip(kmeans.cluster_centers_[0], kmeans.cluster_centers_[1]))
cluster_dict = dict(zip(df.drop('Private', axis=1).columns, cluster_zip))
cluster_df = pd.DataFrame(data=cluster_dict)
cluster_df['Cluster Number'] = cluster_df.index
cluster_df
```

```python
df['Predicted Cluster'] = kmeans.predict(df.drop('Private', axis=1).values)
```

```python
plt.figure(figsize= (10,7))
sns.scatterplot(x='Room.Board', y='Grad.Rate', data=df, alpha=0.6, hue='Predicted Cluster', palette={0:'red', 1:'black'})
sns.scatterplot(x='Room.Board', y='Grad.Rate', data=cluster_df, color='red', s=300, marker='x', 
                hue='Cluster Number',
                palette={0:'red', 1:'black'})

#  Graduation rate and Room and board rate are not good predictors of clustering
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='Outstate', y='F.Undergrad', data=df, alpha=0.7, hue ='Predicted Cluster',palette={0:'red', 1:'black'})
sns.scatterplot(x='Outstate', y='F.Undergrad', data=cluster_df, color='red', s=300, marker='x',
                hue='Cluster Number', palette={0:'red', 1:'black'})
plt.title('K Means Clustering of University Enrollment Data Set')
#  number of full time undergrads looks like a good predictor of cluster number
```

## Model Comparison
Comparing k-means clustering to Private vs Public university label. This is not necessarily an evaluation of the model performance, as model was very generally trained to give two group, without regard to private or public.

```python
df['Private'].unique()
```

```python
df['Cluster'] = df.loc[:, 'Private'].apply(lambda x: 1 if x=='Yes' else 0)
df['Cluster'].head(25).values
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_true = df['Cluster'], y_pred=kmeans.labels_))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true=df['Cluster'], y_pred=kmeans.labels_))
```
